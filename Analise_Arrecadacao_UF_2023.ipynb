```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Define o formato de exibição para floats
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df_original = pd.read_csv('Dados/arrecadacao_estado.csv', sep=';', encoding='latin1')
df_original
```

```python
df_UF = df_original.copy()
df_UF_2023 = df_UF.query("Ano in [2023]")
df_UF_2023 = df_UF_2023.dropna(axis=1, how='all')
df_UF_2023
```

```python
data_types = df_UF_2023.dtypes
print(data_types)
```

```python
def converter_valores(df):
    # Crie uma cópia do DataFrame original
    df_copia = df.copy()
    # Seleciona apenas as colunas a partir da quarta posição
    columas_converter = df_copia.columns[3:]
    for x in columas_converter:
        try:
            # Remove os pontos das strings
            df_copia[x] = df_copia[x].str.replace('.', '')
            # Troca as vírgulas por pontos
            df_copia[x] = df_copia[x].str.replace(',', '.')
            # Converte para float
            df_copia[x] = df_copia[x].astype(float)
            # Se todos valores forem inteiros, converte para int
        except ValueError as e:
            # Se a conversão falhar, mantém a coluna como object e imprime o erro
            print(f"Erro ao converter a coluna {x}: {e}")
            continue
    return df_copia

# Convertendo as colunas
df_UF_2023 = converter_valores(df_UF_2023)
```

```python
df_UF_2023
```

```python
desc_df_UF_2023 = df_UF_2023.iloc[:, 3:]
desc_df_UF_2023 = desc_df_UF_2023.describe().round(2)
print(desc_df_UF_2023.head(5))
```

```python
# Valor arrecadado por imposto em cada estado no ano de 2023
Imposto_por_UF = df_UF_2023.groupby('UF').sum(numeric_only=True).iloc[:, 1:]
Imposto_por_UF = Imposto_por_UF.sort_values(by=Imposto_por_UF.columns[0], ascending=False)
Imposto_por_UF
```

```python
# Somando todos os impostos para cada estado
Total_por_UF= Imposto_por_UF.sum(axis=1)
Total_por_UF = Total_por_UF.sort_values(ascending=False)
print(Total_por_UF.head(5))

# Configurando o gráfico de barras
plt.figure(figsize=(14, 8))
bars = plt.bar(Total_por_UF.index, Total_por_UF.values)

# Adicionando rótulos e título
plt.xlabel('Estados')
plt.ylabel('Soma dos Impostos Arrecadados')
plt.title('Total dos Impostos Arrecadados em 2023 por Estado')
plt.xticks(rotation=90)

ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')))

# Exibindo o gráfico
plt.show()
```

```python
# IMPOSTO SOBRE IMPORTAÇÃO E EXPORTAÇÃO
II_IE = Imposto_por_UF.iloc[:, 0:2]
print(II_IE.head(5))

# Agrupar por estado (UF) e somar os valores
II_IE_por_estado = II_IE.groupby('UF').sum()

# Somar os valores de importação e exportação para cada estado
II_IE_por_estado['Total II/IE'] = II_IE_por_estado['IMPOSTO SOBRE IMPORTAÇÃO'] + II_IE_por_estado['IMPOSTO SOBRE EXPORTAÇÃO']

# Ordenar os estados pelo valor total de II_IE (maior para menor)
II_IE_por_estado = II_IE_por_estado.sort_values(by='Total II/IE', ascending=False)

# Configurando o gráfico de barras
plt.figure(figsize=(14, 8))
bars = plt.bar(II_IE_por_estado.index, II_IE_por_estado['Total II/IE'])

# Adicionando rótulos e título
plt.xlabel('Estados')
plt.ylabel('Soma dos Impostos sobre Importação e Exportação')
plt.title('Total dos Impostos sobre Importação e Exportação em 2023 por Estado')
plt.xticks(rotation=90)

# Formatando os valores do eixo y para serem exibidos com separador de milhares e duas casas decimais
ax = plt.gca()
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')))

# Exibindo o gráfico
plt.show()
```
